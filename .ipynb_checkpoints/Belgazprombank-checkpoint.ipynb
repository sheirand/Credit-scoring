## 2.1 Изучение целевое переменной Marker

```python
#нарисуем график для наглядности
target_plot = df['Marker'].value_counts(normalize=True).mul(100).round(2)
target_plot.index = ['Хороший заёмщик', 'Плохой заёмщик']

plt.figure(figsize=(12,7))
ax = sns.barplot(x = target_plot.index, y= target_plot.values)


for i, p in zip([0,1], ax.patches):
    percentage = target_plot.values[i]
    ax.annotate(f"{target_plot.values[i]}%", 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center',
                va='center',
                xytext=(0,10),
                textcoords='offset points',
                fontsize=14)
plt.title('Распределение целевой переменной\n', fontsize=18)
plt.xlabel('Статус заёмщика', fontsize=14)
plt.ylabel('% в выборке', fontsize=14)
plt.show()
```

# 4. Модель

## 4.0. Подготовка данных

```python
#cat_cols.remove('BIRTH CNTR')
#df.drop(['PROFESSION', 'BIRTH CNTR'], axis=1, inplace=True)
# for i in ['Age', 'DURATION LAST OVERDUE', 'COUNTGUARANTEE',
#           'DURATIONGREATESTDELAYLASTYEAR','DURATIONGREATESTDELAYLASTALL','Loans total']:
#     df[i] = np.log(df[i]+1)
cat_cols.remove('PROFESSION')
df.drop(['PROFESSION'], axis=1, inplace=True)
df_enc = pd.get_dummies(df, columns=cat_cols, drop_first=True)
df_enc.head()
```

```python
df_enc.info()
```

```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = df_enc.drop('Marker', axis=1)
y = np.array(df_enc.Marker)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=RANDOM_STATE, stratify=y)
```

```python
num_cols = [item for item in df.columns if item not in (binary_cols+cat_cols+['Marker'])]
X_train[num_cols[:-1]] = scaler.fit_transform(X_train[num_cols[:-1]])
X_test[num_cols[:-1]] = scaler.transform(X_test[num_cols[:-1]])
X_train.head()
```

```python
#Эксперимент - понижение размерности
from sklearn.decomposition import PCA
pca_test = PCA(n_components=55)
pca_test.fit(X_train)
sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.axvline(linewidth=4, color='r', linestyle = '--', x=30, ymin=0, ymax=1)
display(plt.show())
evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cvr
pca_df['Explained Variance Ratio'] = evr
display(pca_df)
```

## 4.1. Логистическая регрессия 

```python
#clf = LogisticRegression(class_weight='balanced')
#clf = LogisticRegression(solver='liblinear', penalty='l1', random_state=RANDOM_STATE, class_weight='balanced')
clf = RandomForestClassifier(class_weight='balanced')
#clf = XGBClassifier(random_state=RANDOM_STATE, scale_pos_weight=20)
# clf = MultinomialNB()
#clf = GaussianNB()
# clf = BernoulliNB()
#clf = LogisticRegressionCV(cv=3, class_weight='balanced', scoring='recall')
```

```python
def train_eval_model(classifier):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    y_proba = classifier.predict_proba(X_test)[:,1]
    roc_auc = roc_auc_score(y_pred, y_test)
    print(f"train {roc_auc_score(y_train, classifier.predict(X_train))}")
    print(classification_report(y_pred, y_test))
    print('='*55)
    print(f"ROC-AUC score = {roc_auc}")
    print('='*55)
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label='model')
    plt.plot([0, 1], [0, 1], '--', color = 'grey', label = 'random')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve')
    plt.legend(loc = "lower right")
```

```python
train_eval_model(clf)
```
