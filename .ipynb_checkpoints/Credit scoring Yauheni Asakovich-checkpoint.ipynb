# 4. Модель

## 4.0. Подготовка данных

```python
#закодируем категориальные переменные
df_enc = pd.get_dummies(df, columns=cat_cols, drop_first=True)
df_enc.head()
```

```python
#Разделим пространство признаков и ответов на X и y
X = df_enc.drop('Marker', axis=1)
y = np.array(df_enc.Marker)
```

```python
#Разобьем выборку на обучение и тест (стратифицируем ее, т.к. класс 1 присутствует в выборке достаточно редко)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=RANDOM_STATE, stratify=y)
```

```python
#Попробуем использовать PCA
from sklearn.decomposition import PCA
pca_test = PCA(n_components=55)
pca_test.fit(X_train)
sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.axvline(linewidth=4, color='r', linestyle = '--', x=30, ymin=0, ymax=1)
display(plt.show())
evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cvr
pca_df['Explained Variance Ratio'] = evr
display(pca_df)
```

```python
#PCA понизил качество классификации
pca = PCA(n_components=30)
pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
#преобразуем значения числовых переменных с помощью StandardScaler()
scaler = StandardScaler()
#scaler = MinMaxScaler()
#scaler = RobustScaler()
num_cols = [item for item in df.columns if item not in (binary_cols+cat_cols+['Marker'])]
X_train[num_cols[:-1]] = scaler.fit_transform(X_train[num_cols[:-1]])
X_test[num_cols[:-1]] = scaler.transform(X_test[num_cols[:-1]])
X_train.head()
```

## 4.1. Модель. Обучение и оценка

```python
#выберем модель
clf = LogisticRegression(class_weight='balanced', random_state=RANDOM_STATE)
#clf = LogisticRegression(solver='liblinear', penalty='l1', random_state=RANDOM_STATE, class_weight='balanced')
#clf = RandomForestClassifier()
#clf = XGBClassifier()
# clf = MultinomialNB()
#clf = GaussianNB()
#clf = BernoulliNB()
#clf = LogisticRegressionCV(cv=3, class_weight='balanced', scoring='recall')
#clf = SVC(probability=True, class_weight='balanced')
```

```python
def train_eval_model(classifier):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    y_proba = classifier.predict_proba(X_test)[:,1]
    roc_auc = roc_auc_score(y_pred, y_test)
    print(f" Train ROC-AUC score = {roc_auc_score(y_train, classifier.predict(X_train))}")
    print(classification_report(y_pred, y_test))
    print('='*55)
    print(f" Test ROC-AUC score = {roc_auc}")
    print('='*55)
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label='model')
    plt.plot([0, 1], [0, 1], '--', color = 'grey', label = 'random')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve')
    plt.legend(loc = "lower right")
```

```python
#обучим модель и выведем основные метрики качества
train_eval_model(clf)
```
